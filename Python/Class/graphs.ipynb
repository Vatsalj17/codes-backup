```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y1=[20,40,30,10,60]
y2=[10,20,40,50,20]
y3=[30,20,100,40,10]
y4=[40,50,10,60,70]
plt.figure(figsize=(10,5), facecolor='grey')
#plt.gca().set_facecolor('lightgreen')
plt.plot(x,y1, color='b', marker='o', lw=2, label='Royal Enfield')
plt.plot(x,y2, color='r', marker='o', lw=2, label='Himalayn')
plt.plot(x,y3, color='b', marker='o', lw=2, label='Harley')
plt.plot(x,y4, color='m', marker='o', lw=2, label='Honda')

plt.xticks(np.arange(1,6,step=1))
plt.yticks(np.arange(0,210,step=20))
plt.legend()
plt.grid()
plt.title("Bike Details")
plt.xlabel("Number of Days")
plt.ylabel("Distance covered in KMs")
plt.show()
```
